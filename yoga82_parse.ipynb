```python
import os
import requests

# Function to download images
def parse_dataset(file_path, save_folder):
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)
    with open(file_path, 'r') as file:
        lines = file.readlines()

    for line in lines:
        # Split each line into components
        components = line.strip().split('\t')
        #print(components)
        # Extract image address and name
        image_address_name = components[0]
        # Construct the full URL for the image
        full_url = components[1]
        
        # Create the save path for the image
        image_name = image_address_name.split('/')[-1]
        save_path = os.path.join(save_folder, image_name)
        
        # Download and save the image
        download_image(full_url, save_path)

# Function to download a single image
def download_image(url, save_path):
    if os.path.exists(save_path):
        print(f"Image already exists: {save_path}")
        return
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()

        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
        
        print(f"Image saved: {save_path}")

    except requests.exceptions.RequestException as e:
        print(f"Error downloading image from {url}: {e}")



```
